```python
#Remove £ sign in each value
df['Cost']=df['COST (£) charged to Wellcome (inc VAT when charged)'].str.replace('£','')
df['Cost']=df['Cost'].str.replace('$','')
df['Cost']=df['Cost'].str.replace(' ','')
df['Cost']=pd.to_numeric(df['Cost'], errors='coerce')
```

```python
df_top5=df.loc[(df['Journal title']=='plos one') | (df['Journal title']=='journal of biological chemistry') | (df['Journal title']=='proceedings of the national academy of sciences') | (df['Journal title']=='neuroimage') | (df['Journal title']=='nucleic acids research'),['Journal title','Cost']]
print('Cost mean for each top 5 journals')
print(df_top5.groupby('Journal title').mean())
```

```python
print('Cost median for each top 5 journals')
print(df_top5.groupby('Journal title').median())
```

```python
print('Cost standard deviation for each top 5 journals')
print(df_top5.groupby('Journal title').std())
```
